# Pobranie opini o produkcie

```python
response = requests.get("https://www.ceneo.pl/124893467#tab=reviews", headers=headers)  # Replace with the appropriate URL
page_dom = BeautifulSoup(response.text, "html.parser")
opinions = page_dom.select("div.js_product-review:not(.user-post--highlight)")
```

```python
print(f"Found {len(opinions)} opinions on the page.")
```

# wyciągniecie dla każdej opinii jej składowych

```python
all_opinions = []
for opinion in opinions:
    single_opinion = {
        'opinion_id': opinion["data-entry-id"],
        'author': opinion.select_one("span.user-post__author-name").text.strip(),
        'recommendation': opinion.select_one("span.user-post__author-recomendation > em").text.strip() if opinion.select_one("span.user-post__author-recomendation > em") else None,
        'rating': opinion.select_one("span.user-post__score-count").text.strip(),
        'content': opinion.select_one("div.user-post__text").text.strip(),
        'pros': [pro.text.strip() for pro in opinion.select("div.review-feature__item--positive")],
        'cons': [con.text.strip() for con in opinion.select("div.review-feature__item--negative")],
        'helpful': opinion.select_one("button.vote-yes > span").text.strip(),
        'unhelpful': opinion.select_one("button.vote-no > span").text.strip(),
        'publish_date': opinion.select_one("span.user-post__published > time:nth-child(1)")['datetime'].strip(),
        'purchase_date': opinion.select_one("span.user-post__published > time:nth-child(2)")['datetime'].strip() if opinion.select_one("span.user-post__published > time:nth-child(2)") else None
    }
    all_opinions.append(single_opinion)
```

```python
import json

# Zapis danych do pliku JSON
with open("opinions.json", "w", encoding="utf-8") as file:
    json.dump(all_opinions, file, ensure_ascii=False, indent=4)

print("Opinions saved to opinions.json")
```
